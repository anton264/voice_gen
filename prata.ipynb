```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio
from scipy.io.wavfile import write as write_wav
import os
from pydub import AudioSegment
from pydub.silence import detect_nonsilent

# download and load all models
preload_models()

def remove_silence_from_audio(audio, output_file_path, silence_threshold=-50.0, min_silence_len=100):
    non_silent_ranges = detect_nonsilent(audio, min_silence_len, silence_threshold)

    if len(non_silent_ranges) == 0:
        print(f"No audio found. Skipping.")
        return

    start_trim = non_silent_ranges[0][0]
    end_trim = non_silent_ranges[-1][1]
    trimmed_audio = audio[start_trim:end_trim]
    trimmed_audio.export(output_file_path, format="wav")
```

```python
# generate audio from text
text_prompt = """
       The car behind is: Jim
"""
audio_array = generate_audio(text_prompt, history_prompt="en_speaker_6", waveform_temp=0.1)
write_wav("audio.wav", SAMPLE_RATE, audio_array)

# play text in notebook
# Audio(audio_array, rate=SAMPLE_RATE)
```
